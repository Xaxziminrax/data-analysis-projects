```python
students['course_grade'] = students["course_grade"].replace("\"N/A\"", np.nan)
```

```python
students = students.drop_duplicates()
students
```

```python
students["first_name"] = students["first_name"].str.strip().str.upper()
students
```

```python
students["last_name"] = students["last_name"].str.strip().str.upper()
students
```

```python
students["major"] = students["major"].str.strip().str.title()
students["major"] = students["major"].replace({
"Comp Sci" : "Computer Science",
"Cs" : "Computer Science",
"Bio" : "Biology",
"Psych" : "Psychology"
})
```

```python
students["major"].value_counts()
```

```python
students["course_grade"].value_counts()
```

```python
students["course_grade"] = students["course_grade"].str.strip()
students.loc[students["course_grade"] == "ninety", "course_grade"] = 90
```

```python
students["course_grade"].value_counts()
```

```python
students["age"] = pd.to_numeric(students['age'], errors = "coerce")
```

```python
students['age'].describe()
```

```python
students.loc[students["age"] < 5, "age"] = np.nan
students.loc[students["age"] > 120, "age"] = np.nan
```

```python
students["age"].describe()
```

```python
students["age"].dropna().sort_values(ascending=False)
```

```python
#create a column combining first and last names
students["full_name"] = students["last_name"] + ', ' + students["first_name"]
students["full_name"].sort_values()
```

```python
students["age_group"] = pd.cut(
    students["age"],
    bins=[0, 25, 65, 120], 
    labels=["Young Adults", "Working Adults", "Retired"]
)
```

```python
students[["age", "age_group"]].value_counts()
```

```python
students["costPerSemester"] = students['tuition_paid'] / 3
students["costPerSemester"].sort_values(ascending=False).dropna()
```

```python
students["costPerSemester"].describe()
```

```python
students["enrollment_date"] = pd.to_datetime(students["enrollment_date"])
```
